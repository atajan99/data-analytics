## Breast Cancer and Decision Tree
Our sample data is data of images collected from
breast cancer cells. Features are computed from a digitized image of a fine needle
aspirate (FNA) of a breast mass. They describe the characteristics of the cell nuclei
present in the image. Diagnoses are shown as a class. We have two classes (2 for
malignant and 4 for benign).

<b> Let's print accuracy of our model and print the visualized decision tree of our model. </b>

**Load Packages**

Start with loading required packages

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
```

**Loading Data**

Load the data here

```python
#import os
#os.getcwd()
df = pd.read_csv("breast_cancer_dataset.csv") #put the adress to file here
```

It is always recommanded to get some sense about the data. Let's look into the data more closely.

```python
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

**0- Identify X and y**

Assign y to the “class” variable and X to the rest of the variables.

```python
y= pd.DataFrame(df['class']) #assign the colun class to y
X= df.drop('class', axis=1)
```

**1-Data Split**

Split the data into 80% for training and 20% for test

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.20, random_state=52) #write the appropriate code here. Use random_state=1
```

**2- Model Design**

Create a decision tree model using training data

```python
model = DecisionTreeClassifier() #Call decision tree classifier
model = model.fit(X_train, y_train) #fit the model on training data
```

```python
#If you like to see some of the specifications of the model, you can type...
model.__dict__
```

**3-Test the model**


Test the model using Test data

```python
y_pred =model.predict(X_test) #Use test data to make prediction
y_pred
#y_test
```

```python
#Confusion Matrix
#Write the code for showing confusion matrix here. The output will look like below
confusion_matrix(y_test, y_pred)
#confusion_matrix()
```

```python
#print acuracy here
#Compuet the accuracy here
#You can multiply this number to measure it in percentage
from sklearn import metrics
import math

print((metrics.accuracy_score(y_test, y_pred))*100)
```

**4- Plot the tree**

Plot the decision tree

```python
# The majority of the code is provided here. You just need to add a few things
import matplotlib.pyplot as plt
from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(model, 
                   feature_names=list(X.columns),  
                   class_names=np.array(y['class'].unique()).astype(str),#['2','4'],
                   filled=True)
```
